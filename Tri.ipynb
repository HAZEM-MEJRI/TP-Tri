 <span style="color:#B37800 "> Selection :

```python
def tri_S(v):
    for i in range(len(v)):
        ind_min = i
        for k in range(i+1,len(v)):
            if (v[k] < v[ind_min]):
                ind_min = k
        aux = v[i]
        v[i]= v[ind_min]
        v[ind_min]=aux
    e = datetime.now() 
    diff = e-s 
    sec = diff.total_seconds() 
    tmp_selection.append(sec)
```

 <span style="color:#B37800 "> Insertion :

```python
def tri_Inser (v):
    p=len(v)-1
    for i in range(0,p) :
        x = v[i+1]
        j=i
        while((v[j] > x ) and (j>=0) ) :
            v[j+1] = v[j]
            j-= 1
        v[j+1] = x
    e = datetime.now() 
    diff = e-s 
    sec = diff.total_seconds() 
    tmp_insertion.append(sec)
```
